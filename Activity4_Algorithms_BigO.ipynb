```python
# Algorithm to find the Smallest number in a List
numbers = [10, 5, 20, 8, 15]
smallest = numbers[0]  
for num in numbers:
    if num < smallest: 
        smallest = num 

print("Smallest number:", smallest)
```

```python
def print_pairs(items):
    for i in items:
        for j in items:
            print(i, j)

def print_first(items):
    if len(items) > 0:
        print(items[0])

sample = [1, 2, 3]
print_pairs(sample)
print_first(sample)
```

```python
def linear_search(arr, target):
    comparisons = 0
    for i in range(len(arr)):
        comparisons += 1
        if arr[i] == target:
            return i, comparisons
    return -1, comparisons

data = [3, 5, 2, 4, 9]
index, comps = linear_search(data, 4)
print("Found at index:", index)
print("Comparisons:", comps)
```

```python
def binary_search(arr, target):
    low = 0
    high = len(arr) - 1
    comparisons = 0

    while low <= high:
        comparisons += 1
        mid = (low + high) // 2

        if arr[mid] == target:
            return mid, comparisons
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return -1, comparisons

sorted_data = [1, 3, 5, 7, 9, 11]
result, comps = binary_search(sorted_data, 7)

print("Found at index:", result)
print("Comparisons:", comps)
```

```python
def bubble_sort(arr):
    n = len(arr)
    swaps = 0

    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                swaps += 1
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

    return arr, swaps

to_sort = [64, 34, 25, 12, 22, 11, 90]
sorted_arr, swaps = bubble_sort(to_sort)

print("Sorted list:", sorted_arr)
print("Swaps:", swaps)
```

```python
def selection_sort(arr):
    comparisons = 0
    swaps = 0

    for i in range(len(arr)):
        min_idx = i
        for j in range(i+1, len(arr)):
            comparisons += 1
            if arr[j] < arr[min_idx]:
                min_idx = j
        swaps += 1
        arr[i], arr[min_idx] = arr[min_idx], arr[i]

    return arr, comparisons, swaps

bubble_arr, bubble_swaps = bubble_sort([5, 3, 8, 4, 2])
selection_arr, sel_comp, sel_swaps = selection_sort([5, 3, 8, 4, 2])

print("Bubble Swaps:", bubble_swaps)
print("Selection Comparisons:", sel_comp)
print("Selection Swaps:", sel_swaps)
```

```python
def insertion_sort(arr):
    comparisons = 0

    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1

        while j >= 0 and key < arr[j]:
            comparisons += 1
            arr[j + 1] = arr[j]
            j -= 1

        arr[j + 1] = key

    return arr, comparisons

arr = [1, 2, 3, 4, 5]  # already sorted
sorted_arr, comps = insertion_sort(arr)

print("Sorted array:", sorted_arr)
print("Comparisons:", comps)
```
