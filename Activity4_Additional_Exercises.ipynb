```python
#Exercise 1
import time

# Inputs
student_id = 253625
main_list = list(range(10000)) # Hardcoded sorted list for searching
small_list = [1, 5, 8, 3, 9, 2, 8, 4] # Hardcoded small list with duplicate '8'

# Calculations
target = int(str(student_id)[-2:]) # Extracts '56'

# 1. Linear Search O(n)
start_linear = time.perf_counter()
linear_found = False
for num in main_list:
    if num == target:
        linear_found = True
        break
time_linear = time.perf_counter() - start_linear

# 2. Binary Search O(log n)
start_binary = time.perf_counter()
low, high = 0, len(main_list) - 1
binary_found = False
while low <= high:
    mid = (low + high) // 2
    if main_list[mid] == target:
        binary_found = True
        break
    elif main_list[mid] < target:
        low = mid + 1
    else:
        high = mid - 1
time_binary = time.perf_counter() - start_binary

# 3. Duplicate Search O(n^2)
start_quad = time.perf_counter()
duplicate_found = None
for i in range(len(small_list)):
    for j in range(i + 1, len(small_list)):
        if small_list[i] == small_list[j]:
            duplicate_found = small_list[i]
            break
time_quad = time.perf_counter() - start_quad

# Output Table
print(f"Target Value: {target}")
print("-" * 55)
print(f"{'Algorithm':<20} | {'Complexity':<10} | {'Time (Seconds)':<15}")
print("-" * 55)
print(f"{'Linear Search':<20} | {'O(n)':<10} | {time_linear:.8f}")
print(f"{'Binary Search':<20} | {'O(log n)':<10} | {time_binary:.8f}")
print(f"{'Duplicate Search':<20} | {'O(n^2)':<10} | {time_quad:.8f}")
print("-" * 55)
```

```python
#Exercise 2
import time

# Input
student_id = 253625

# Calculations
last_digit = int(str(student_id)[-1]) # Extracts '6'
n = last_digit + 15 # n = 21

def fib_iterative(n):
    if n <= 1:
        return n
    a, b = 0, 1
    for _ in range(2, n + 1):
        a, b = b, a + b
    return b

start_time = time.perf_counter()
result = fib_iterative(n)
exec_time = time.perf_counter() - start_time

# Output Table
print("-" * 45)
print(f"{'Problem Size (n)':<18} | {'Result':<10} | {'Time (s)':<10}")
print("-" * 45)
print(f"{n:<18} | {result:<10} | {exec_time:.8f}")
print("-" * 45)
```

```python
#Exercise 3
import time
import random

# Input & Array Sizing
student_id = 253625
last_three_digits = int(str(student_id)[-3:]) # Extracts '456'
list_size = last_three_digits + 50 # Size = 506

# Generate master list
base_list = random.sample(range(10000), list_size)

# Sorting Functions
def bubble_sort(arr):
    n = len(arr)
    comparisons, swaps = 0, 0
    for i in range(n):
        for j in range(0, n - i - 1):
            comparisons += 1
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swaps += 1
    return comparisons, swaps

def selection_sort(arr):
    n = len(arr)
    comparisons, swaps = 0, 0
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            comparisons += 1
            if arr[j] < arr[min_idx]:
                min_idx = j
        if min_idx != i:
            arr[i], arr[min_idx] = arr[min_idx], arr[i]
            swaps += 1
    return comparisons, swaps

def insertion_sort(arr):
    comparisons, swaps = 0, 0
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0:
            comparisons += 1
            if arr[j] > key:
                arr[j + 1] = arr[j]
                swaps += 1
                j -= 1
            else:
                break
        arr[j + 1] = key
    return comparisons, swaps

# Execution & Timing
# Bubble
arr1 = base_list[:]
start = time.perf_counter()
b_comp, b_swaps = bubble_sort(arr1)
b_time = time.perf_counter() - start

# Selection
arr2 = base_list[:]
start = time.perf_counter()
s_comp, s_swaps = selection_sort(arr2)
s_time = time.perf_counter() - start

# Insertion
arr3 = base_list[:]
start = time.perf_counter()
i_comp, i_swaps = insertion_sort(arr3)
i_time = time.perf_counter() - start

# Output Table
print(f"Sorting List Size: {list_size} elements")
print("-" * 65)
print(f"{'Algorithm':<15} | {'Comparisons':<12} | {'Swaps':<10} | {'Time (Seconds)'}")
print("-" * 65)
print(f"{'Bubble Sort':<15} | {b_comp:<12} | {b_swaps:<10} | {b_time:.5f}")
print(f"{'Selection Sort':<15} | {s_comp:<12} | {s_swaps:<10} | {s_time:.5f}")
print(f"{'Insertion Sort':<15} | {i_comp:<12} | {i_swaps:<10} | {i_time:.5f}")
print("-" * 65)
```
